```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
url = 'gs://vcm-ml-intermediate/2024-09-04-era5-1deg-8layer-forcing-clim-1991-2020.zarr'  # this is generated by compute.py script in this directory
times = ['2001-01-15T00:00:00', '2001-04-15T00:00:00', '2001-07-15T00:00:00', '2001-10-15T00:00:00']
```

```python
ds = xr.open_zarr(url)
```

```python
ds.DSWRFtoa.sel(latitude=80, longitude=180, method='nearest').plot()
```

```python
ds.sea_ice_fraction.sel(latitude=80, longitude=180, method='nearest').plot()
```

```python
ds.surface_temperature.sel(latitude=45, longitude=340, method='nearest').plot()
```

```python
for name in ds.data_vars:
    if 'latitude' in ds[name].dims:
        plt.figure()
        if 'time' in ds[name].dims:
            ds[name].sel(time=times).plot(col='time', col_wrap=2)
        else:
            ds[name].plot()
```
